```r
set.seed(123)
populacao = rexp(10000, 1/5)
hist(populacao)
```

```r
amostras = 50
n_amostra = 6
```

```r

media = numeric(amostras) # inicializando o vetor

for (i in 1:amostras){
  amostra = sample(populacao, n_amostra, replace = T)
  media[i] = mean(amostra)
}
```

```r
hist(media)
qqnorm(media)
qqline(media)
shapiro.test(media)
```
